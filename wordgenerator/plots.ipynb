```python
data =[
		185,
		191,
		246,
		199,
		240,
		229,
		211,
		205,
		207,
		207,
		227,
		212,
		234,
		197,
		224,
		208,
		194,
		221,
		211,
		202,
		206,
		210,
		218,
		185,
		205,
		195,
		195,
		198,
		203,
		183,
		191,
		211,
		201,
		217,
		207]
sections  = [f" 10k " for i in range(1,36)]
print(sections)
print(data)
print(len(sections))
print(len(data))
```

```python
import matplotlib.pyplot as plt
import numpy as np


# Rastgele renkler
colors = np.random.rand(len(data), 3)

plt.figure(figsize=(10,6))
plt.bar(np.arange(len(data)), data, color=colors)

plt.xticks(np.arange(len(data)), [f"10k" for i in range(len(data))], rotation=90)
plt.title("Correctly Guessed password numbers for every 10k generations")
plt.ylabel("10ks")
plt.xlabel("Number of password batch")
plt.tight_layout()
plt.show()
```

```python
data3 = np.array(data)
data3 = data3 / 10000
data3 = data3*100

plt.figure(figsize=(10,6))
plt.bar(np.arange(len(data3)), data3, color=colors)

plt.xticks(np.arange(len(data3)), [f"10k" for i in range(len(data))], rotation=90)
plt.title("Hit Rate in every 10k")
plt.ylabel("hitrate %")
plt.xlabel("10ks")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10,6))
data2  = np.array(data)
data2 = np.cumsum(data2)
plt.plot(np.arange(len(data2)),data2,'o-')
plt.xticks(np.arange(len(data2)), [f"{10*(i+1)}k" for i in range(len(data))], rotation=45)
plt.title("Correctly Guessed passwords")
plt.ylabel("10ks")
plt.xlabel("Number of generated password")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10,6))
data2  = np.array(data)
data2 = np.cumsum(data2)
data2=data2/int(1e+6)
data2 = data2*100
plt.plot(np.arange(len(data2)),data2,'o-')
plt.xticks(np.arange(len(data2)), [f"{10*(i+1)}k" for i in range(len(data))], rotation=45)
plt.title("Correctly Guessed passwords percentage")
plt.ylabel("percentage")
plt.xlabel("Number of generated password")
plt.tight_layout()
plt.show()
```
