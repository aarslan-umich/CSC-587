```python
import sklearn
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
import pandas as panda
```

```python
#read data into file ans split without using sklearn spilitter
data =[]
feature_name=None 
with open("diabetes_train.csv","r+") as myfile:
    counter=0
    for line in myfile:
        if counter==0:
            feature_name = line.strip().split(",")
            counter +=1
            continue
        data.append(line.strip().split(","))
train_data = data[:-10]
test_data = data[-10:]
train_data = panda.DataFrame(train_data,columns=feature_name)
test_data = panda.DataFrame(test_data,columns=feature_name)

train_Y = train_data["class"]
train_data = train_data.drop("class",axis=1)

test_Y = test_data["class"]
test_data = test_data.drop("class",axis=1)

```

```python
#scale Data
scaler = StandardScaler()
train_data = scaler.fit_transform(train_data)
test_data = scaler.transform(test_data)

#label encoding for targets
labels = {"tested_positive": 1 , "tested_negative": 0 }
train_Y = train_Y.map(labels)
test_Y= test_Y.map(labels)
```

```python
svm_classifier = SVC(kernel='rbf', C=10,class_weight='balanced')
svm_classifier.fit(train_data,train_Y)
prediction  = svm_classifier.predict(test_data)
print(f"Accoracy : {accuracy_score(test_Y,prediction)} - precision : {precision_score(test_Y,prediction)} \n recall : {recall_score(test_Y,prediction)} - F1 : {f1_score(test_Y,prediction)}")
```
